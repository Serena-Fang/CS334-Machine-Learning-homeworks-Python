```python
import argparse
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def model_assessment(filename):
    """
    Given the entire data, split it into training and test set 
    so you can assess your different models 
    to compare perceptron, logistic regression,
    and naive bayes. 
    """
    label = []
    email = []
    with open('spamAssassin.data') as file:
        lines = file.readlines()
    for line in lines:
        label.append(line[:1])
        email.append(line[2:-1])
    x_train, x_test, y_train, y_test = train_test_split(email, label, test_size=0.3, random_state=1)
    return x_train, x_test, y_train, y_test
```

```python
data =  pd.read_csv('spamAssassin.data', sep=",", header = None)
label = {'label': data[0].str[:1]}
email = {'email': data[0].str[2:]}
df_x = pd.DataFrame(email)
df_y = pd.DataFrame(label)
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=1)
```

```python
def build_vocab_map():
    cv = CountVectorizer()
    x_traincv = cv.fit_transform(df_x['email'])
    count = x_traincv.toarray()
    vocab = cv.get_feature_names_out()
    df1 = pd.DataFrame(count, columns=vocab)
    word_sum = df1.sum(axis=0)
    map3 = pd.DataFrame(word_sum)
    map4 = map3.rename(columns={0: 'count'})
    vocab_map = map4.loc[map4['count'] >= 30]
    return vocab_map.head()
```

```python
def construct_count():
    """
    Construct email datasets based on
    the count representation of the email.
    For each e-mail, transform it into a
    feature vector where the ith entry,
    $x_i$, is the number of times the ith word in the 
    vocabulary occurs in the email,
    or 0 otherwise
    """
    return None
```

```python
def main():
    """
    Main file to run from the command line.
    """
    # set up the program to take in arguments from the command line
    parser = argparse.ArgumentParser()
    parser.add_argument("--data",
                        default="spamAssassin.data",
                        help="filename of the input data")
    args = parser.parse_args()
    model_assessment(args.data)
    build_vocab_map()
    construct_binary()
    construct_count()
```
